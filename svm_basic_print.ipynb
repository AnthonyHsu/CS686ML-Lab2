```python
from svm_basic import svm_basic
```

```python
svm = svm_basic()
dataArr, labelArr = svm.loadDataSet('linearly_separable.csv')
splitIndex = int(len(dataArr) * 0.9)

train_data = dataArr[:splitIndex]
train_label = labelArr[:splitIndex]
test_data = dataArr[splitIndex:]
test_label = labelArr[splitIndex:]

b, alphs, ws = svm.fit(train_data, train_label)
wslist = [b.getA()[0][0], ws[0][0], ws[1][0]]

svm.plot_fit(wslist, train_data, train_label)
```

```python
hypo = svm.predict(test_data, wslist)
print('Accuracy is:', svm.accuracy(test_label, hypo))
svm.print_confusion_matrix(test_label, hypo)
```
